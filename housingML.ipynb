```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\ASUS VIVOBOOK\OneDrive\Desktop\python code\Housing.csv")
print(data)
data.dtypes
```

```python
data["mainroad"]=data["mainroad"].astype("category")
data["guestroom"]=data["guestroom"].astype("category")
data["basement"]=data["basement"].astype("category")
data["hotwaterheating"]=data["hotwaterheating"].astype("category")
data["airconditioning"]=data["airconditioning"].astype("category")
data["furnishingstatus"]=data["furnishingstatus"].astype("category")
data["prefarea"]=data["prefarea"].astype("category")
```

```python
catdata=data.drop(['price','area','bedrooms','stories','parking','bathrooms'],axis=1)
print(catdata)
```

```python
catdata['hotwaterheating']=catdata['hotwaterheating'].cat.codes
catdata['mainroad']=catdata['mainroad'].cat.codes
catdata['guestroom']=catdata['guestroom'].cat.codes
catdata['basement']=catdata['basement'].cat.codes
catdata['airconditioning']=catdata['airconditioning'].cat.codes
catdata['prefarea']=catdata['prefarea'].cat.codes
catdata['furnishingstatus']=catdata['furnishingstatus'].cat.codes
print(catdata)
```

```python
numdata=data.drop(['guestroom','basement','hotwaterheating','airconditioning','furnishingstatus','prefarea','mainroad'],axis=1)
print(numdata)
```

```python
newdata=numdata.join(catdata)
print(newdata)
```

```python
data1=newdata.values
data1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
newtarget=data1[:,0]
newdata=data1[:,1:]
traindata,testdata,traintarget,testtarget=train_test_split(newdata,newtarget,test_size=0.1)
```

```python
model=LinearRegression()
model.fit(traindata,traintarget)
```

```python
predict=model.predict(testdata)

from sklearn.metrics import r2_score

r2=r2_score(testtarget,predict)
print("r2 score:",r2)
```
